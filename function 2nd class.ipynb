```python
def cal():
    a=int(input("enter a= "))
    b=int(input("enter b= "))
    o=input("enter the */-+ = ")
    if o=="+":
        print(a+b)
    elif o=="-":
        print(a-b)
    elif o=="*":
        print(a*b)
    elif o=="/":
        print(a/b)
    else:
        print("invalid operator")
    
    x=input("enter y if u want more cal = ")
    if x=='y':
        cal()
cal()
```

```python
def add(x,y):
    return x+y
def sub(x,y):
    return x-y
def mul(x,y):
    return x*y
def div(x,y):
    return x/y

def cal():
    a=int(input("enter a= "))
    b=int(input("enter b= "))
    o=input("enter the */-+ = ")
    if o=="+":
        print(add(a,b))
    elif o=="-":
        print(sub(a,b))
    elif o=="*":
        print(mul(a,b))
    elif o=="/":
        print(div(a,b))
        
    else:
        print("invalid operator")
    
    x=input("enter y if u want more cal = ")
    if x=='y':
        cal()
cal()
```
